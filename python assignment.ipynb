```python
data = [
    [120, 1, 'Soumita M', 'Winning Culture Lab'],
    [280, 13, 'Subhangi 0 Nitin Shane', 'Winning Culture Lab'],
    [75, 1124, 'Merwin', 'Mentor'],
    [295, 16, 'fardinkamal62', 'Human Capital Lab'],
    [263, 17, 'riddhi_213', 'Human Capital Lab'],
    [266, 19, 'Rohit Duttal imshawan', 'Mentor Tech Lab'],
    [71, 20, 'Anuraj_Saini', 'Brand Tech Lab'],
    [3367, 22, 'sharath Ronak 0', 'Student Unicorn Lab Tech Lab'],
    [336, 24, 'Amrit Malviya', 'Brandtech Lab'],
    [271, 25, 'Saurabh', 'Mentor Student Unicorn Lab'],
    [3169, 26, 'darshimalde', 'Student Unicorn Lab'],
    [100, 27, 'Shagun', 'Tech Lab'],
    [3406, 28, 'Ayisha', 'Student Unicorn Lab'],
    [69, 29, 'Palash', 'Growpital'],
    [539, 30, 'raman', 'Growpital'],
    [299, 31, 'Nishant', 'Growpital'],
    [3408, 32, 'Vatsal', 'Kringle'],
    [302, 23, '', ''],
]

# Create a dictionary to store team-wise statements and reasons
team_stats = {}

# Loop through the data and calculate the total statements and reasons for each team
for row in data:
    if row[3] not in team_stats:
        team_stats[row[3]] = {'statements': row[1], 'reasons': row[0]}
    else:
        team_stats[row[3]]['statements'] += row[1]
        team_stats[row[3]]['reasons'] += row[0]

# Calculate the average statements and reasons for each team
for team, stats in team_stats.items():
    stats['avg_statements'] = round(stats['statements'] / len([row for row in data if row[3] == team]), 2)
    stats['avg_reasons'] = round(stats['reasons'] / len([row for row in data if row[3] == team]), 2)

# Sort the teams based on their average statements and reasons
sorted_teams = sorted(team_stats.items(), key=lambda x: (x[1]['avg_statements'], x[1]['avg_reasons']), reverse=True)

# Print the leaderboard
print('Team Rank\n')
print('Thinking Teams Leaderboard\n')
print('Average Statements\tAverage Reasons')
for i, (team, stats) in enumerate(sorted_teams):
    print(f"{i+1}\t{team}\t{stats['avg_statements']}\t\t{stats['avg_reasons']}")
```
